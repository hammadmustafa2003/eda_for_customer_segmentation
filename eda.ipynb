```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
csv_file_path = "preprocessed_customer_data.csv"
customer_df = pd.read_csv(csv_file_path)

# Summary statistics of the dataset
print(customer_df.describe())
```

```python
customer_df.head()
```

```python
# Pairwise correlation heatmap
correlation_matrix = customer_df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Pairwise Correlation Heatmap")
plt.show()
```

It is clear from the graph that data is completely random and there is no relation between the columns

```python
# Gender distribution
plt.figure(figsize=(6, 4))
sns.countplot(x="Gender", data=customer_df)
plt.title("Gender Distribution")
plt.show()
```

```python
# Marital status distribution
plt.figure(figsize=(8, 6))
sns.countplot(x="MaritalStatus", data=customer_df)
plt.title("Marital Status Distribution")
plt.show()
```

```python
# Preferred category distribution
plt.figure(figsize=(8, 6))
sns.countplot(x="PreferredCategory", data=customer_df)
plt.title("Preferred Category Distribution")
plt.show()
```

```python
# Age distribution
plt.figure(figsize=(10, 6))
sns.histplot(customer_df["Age"], bins=20, kde=True)
plt.title("Age Distribution")
plt.show()
```

```python
# Annual income distribution
plt.figure(figsize=(10, 6))
sns.histplot(customer_df["AnnualIncome (USD)"], bins=20, kde=True)
plt.title("Annual Income Distribution")
plt.show()
```

```python
# Total purchases distribution
plt.figure(figsize=(10, 6))
sns.histplot(customer_df["TotalPurchases"], bins=20, kde=True)
plt.title("Total Purchases Distribution")
plt.show()
```

```python
# Preferred category vs. Total purchases
plt.figure(figsize=(10, 6))
sns.boxplot(x="PreferredCategory", y="TotalPurchases", data=customer_df)
plt.title("Preferred Category vs. Total Purchases")
plt.show()
```

```python
plt.scatter(x=customer_df["AnnualIncome (USD)"], y=customer_df["TotalPurchases"])
plt.xlabel("Annual Income (USD)")
plt.ylabel("Total Purchases")
plt.title("Annual Income vs Total Purchases")
plt.show()
```

```python
plt.scatter(x=customer_df["AnnualIncome (USD)"], y=customer_df["Age"])
plt.xlabel("Annual Income (USD)")
plt.ylabel("Age")
plt.title("Annual Income vs Age")
plt.show()
```
